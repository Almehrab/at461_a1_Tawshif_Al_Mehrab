### Dataset Overview

Twitch dataset consists of four related tables- `viewers` table contains demographic and account information for 1,000 viewers. `streamers` table contains metadata for 76 streamers, including category, language, and follower statistics. `streams` table records 1,199 individual broadcasts, while the `sessions` table contains 22,825 viewing sessions and serves as the main behavioural event log
 tables are connected through `viewer_id`, `streamer_id`, and `stream_id`. The sessions table links viewers to specific streams and streamers, allowing behavioural data to be combined with viewer and streamer metadata for analysis

```python
print("Duplicate viewers:", viewers.duplicated().sum())
print("Duplicate streamers:", streamers.duplicated().sum())
print("Duplicate streams:", streams.duplicated().sum())
print("Duplicate sessions:", sessions.duplicated().sum())
```

```python
print("Duplicate viewer_id:", viewers["viewer_id"].duplicated().sum())
print("Duplicate streamer_id:", streamers["streamer_id"].duplicated().sum())
print("Duplicate stream_id:", streams["stream_id"].duplicated().sum())
print("Duplicate session_id:", sessions["session_id"].duplicated().sum())
```
